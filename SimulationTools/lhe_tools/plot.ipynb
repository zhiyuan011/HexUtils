```python
Jeff_xsec = {
    "ss":4.14e-4,
    "dd":2.97e-6,
    "cc":0.0172,
    "uu":8.84e-7,
    "bb":0.111
}
```

```python
for flav in ("uu", "ss", "dd", "cc"):
    xsec = f[f'CrossSection/unweighted_events_{flav}.lhe']['Value'].array()[0]
    scalefac = xsec*1000/Jeff_xsec[flav]
    print(f"For {flav}, scalefactor is {scalefac}")
```
