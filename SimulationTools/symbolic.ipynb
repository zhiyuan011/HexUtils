```python
import numpy as np
import vector
import awkward as ak
import numba as nb
import sympy
```

```python
g1, g2, g4 = sympy.symbols("g_1 g_2 g_4")
mV, mX = sympy.symbols("m_V m_X")
m_1, m_2, c1, c2, phi = sympy.symbols("m_1 m_2 c_1 c_2 phi")
s1 = 1 - c1**2
s2 = 1 - c2**2
VeV = sympy.symbols("nu")
dGammaV = 2.4955 #width of the Z

Af1 = 0.14876764
```

```python
s = (mX**2 - m_1**2 - m_2**2)/2
x = (s/(m_1*m_2))**2 - 1
x
```

```python
a1 = g1*mV**2/mX**2 + (2*s*g2/mX**2)
a2 = -2*g2
a3 = -2*g4
```

```python
x_factor = (m_1*m_2)/(mX**2)
A00 = -mX**2/VeV*(a1*sympy.sqrt(1 + x) + a2*x*x_factor)
APP = mX**2/VeV*(a1 + sympy.I*a3*sympy.sqrt(x)*x_factor)
AMM = mX**2/VeV*(a1 - sympy.I*a3*sympy.sqrt(x)*x_factor)
```

```python
phi_PP = sympy.arg(APP/A00)
phi_MM = sympy.arg(AMM/A00)
phi_PPMM = sympy.arg(APP/AMM)
```

```python
t1 = 4*sympy.Abs(A00)**2*s1**2*s2**2
t2 = sympy.Abs(APP)**2*(1 + 2*Af1*c1 + c1**2)*(1 + 2*Af1*c2 + c2**2)
t3 = sympy.Abs(AMM)**2*(1 - 2*Af1*c1 + c1**2)*(1 - 2*Af1*c2 + c2**2)
t4 = 4*sympy.Abs(A00)*sympy.Abs(APP)*(Af1 + c1)*s1*(Af1 + c2)*s2*sympy.cos(phi + phi_PP)
t5 = 4*sympy.Abs(A00)*sympy.Abs(AMM)*(Af1 - c1)*s1*(Af1 - c2)*s2*sympy.cos(phi - phi_MM)
t6 = 2*sympy.Abs(APP)*sympy.Abs(AMM)*(s1**2)*(s2**2)*sympy.cos(2*phi + phi_PPMM)

res = t1 + t2 + t3 + t4 + t5 + t6
```

```python
integrated = sympy.integrate(res, m_1)
```
